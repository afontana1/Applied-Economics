```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n = 10; smean = 3.0; sstdev = 1.0; alpha = 0.05

dof = n-1
xval = np.linspace(-6,6,1000)
tpdf = stats.t.pdf(xvals,df=n-1)

plt.subplot(131)
plt.plot(xval,tpdf,color='red',lw=2); plt.xlabel('Values'); plt.ylabel('Density'); 
plt.title(r'Standard Students, $t_{\left(\frac{\alpha}{2},dof\right)}$')

tscore = np.round(-1*stats.t.ppf(alpha/2.0, df=n-1),2)

lower_CI = np.round(-1*tscore,2)
upper_CI = np.round(1*tscore,2)

plt.annotate('lower interval = ' + str(lower_CI),[lower_CI-0.1,stats.t.pdf(lower_CI,df = dof)+0.05],rotation=90.0)
plt.annotate('upper interval = ' + str(upper_CI),[upper_CI-0.1,stats.t.pdf(upper_CI,df = dof)+0.05],rotation=90.0)
plt.annotate('$\overline{x} = $' + str(0),[0-0.1,stats.t.pdf(0,df = dof)+0.05],rotation=90.0)
 
plt.fill_between(xval,tpdf,where= xval< lower_CI, color='red',alpha=0.7)
plt.fill_between(xval,tpdf,where= xval> upper_CI, color='red',alpha=0.7)

plt.xlim([-6,6]); plt.ylim([0,1.6])

se = sstdev/np.sqrt(n)

plt.subplot(132)

setpdf = stats.t.pdf(xval,df=n-1,loc=0.0,scale=se)

tscore = np.round(-1*stats.t.ppf(alpha/2.0, df=dof),2)

lower_CI = np.round(-1*tscore*se,2)
upper_CI = np.round(1*tscore*se,2)

plt.annotate('lower interval = ' + str(lower_CI),[lower_CI-0.1,stats.t.pdf(lower_CI,df = dof, scale = sstdev/np.sqrt(n))+0.05],rotation=90.0)
plt.annotate('upper interval = ' + str(upper_CI),[upper_CI-0.1,stats.t.pdf(upper_CI,df = dof, scale = sstdev/np.sqrt(n))+0.05],rotation=90.0)
plt.annotate('$\overline{x} = $' + str(0),[0-0.1,stats.t.pdf(0,df = dof, scale = sstdev/np.sqrt(n))+0.05],rotation=90.0)
 
plt.fill_between(xval,setpdf,where= xval< lower_CI, color='red',alpha=0.7)
plt.fill_between(xval,setpdf,where= xval> upper_CI, color='red',alpha=0.7)

plt.plot(xvals,setpdf,color='red',lw=2); plt.xlabel('Values'); plt.ylabel('Density'); 
plt.title(r'Students, $t_{\left(\frac{\alpha}{2},dof\right)} \times \frac{s_{x}}{\sqrt{n}}$')
plt.xlim([-6,6]); plt.ylim([0,1.6])

msetpdf = stats.t.pdf(xval,df=n-1,loc=smean,scale=se)

tscore = np.round(-1*stats.t.ppf(alpha/2.0, df=dof),2)
se = np.round(sstdev/(np.sqrt(n)),2)

lower_CI = np.round(smean - tscore*se,2)
upper_CI = np.round(smean + tscore*se,2)

plt.subplot(133)
plt.plot(xvals,msetpdf,color='red',lw=2); plt.xlabel('Values'); plt.ylabel('Density'); 
plt.title(r'Students, $\overline{x} \pm t_{\left(\frac{\alpha}{2},dof \right)} \times \frac{s_{x}}{\sqrt{n}}$')
plt.xlim([-6,6]); plt.ylim([0,1.6])

plt.annotate('lower interval = ' + str(lower_CI),[lower_CI-0.1,stats.t.pdf(lower_CI,df = dof, loc = smean, scale = sstdev/np.sqrt(n))+0.05],rotation=90.0)
plt.annotate('upper interval = ' + str(upper_CI),[upper_CI-0.1,stats.t.pdf(upper_CI,df = dof, loc = smean, scale = sstdev/np.sqrt(n))+0.05],rotation=90.0)
plt.annotate('$\overline{x} = $' + str(smean),[smean-0.1,stats.t.pdf(smean,df = dof, loc = smean, scale = sstdev/np.sqrt(n))+0.05],rotation=90.0)
    
plt.fill_between(xval,msetpdf,where= xval< lower_CI, color='red',alpha=0.7)
plt.fill_between(xval,msetpdf,where= xval> upper_CI, color='red',alpha=0.7)

plt.subplots_adjust(left=0.0, bottom=0.0, right=1.5, top=0.6, wspace=0.3, hspace=0.4); plt.show()
```
